# CSV

```python
from io import StringIO,BytesIO
```

```python
data=('col1,col2,col3\n'
            'x,y,1\n'
            'a,b,2\n'
            'c,d,3')
```

```python
pd.read_csv(StringIO(data))
```

```python
## Read from specific columns
df=pd.read_csv(StringIO(data), usecols=lambda x: x.upper() in ['COL1', 'COL3'])
```

```python
df
```

```python
df.to_csv('Test.csv')
```

```python
## Specifying columns data types

data = ('a,b,c,d\n'
            '1,2,3,4\n'
            '5,6,7,8\n'
            '9,10,11')
```

```python
print(data)
```

```python
df=pd.read_csv(StringIO(data),dtype=object)
```

```python
df
```

```python
df=pd.read_csv(StringIO(data),dtype={'b':int,'c':np.float,'d':'Int64'})
```

```python
df
```

```python
df.dtypes
```

```python
data = ('index,a,b,c\n'
           '4,apple,bat,5.7\n'
            '8,orange,cow,10')
```

```python
## Quoting and Escape Characters¶. Very useful in NLP

data = 'a,b\n"hello, \\"Bob\\", nice to see you",5'
```

```python
pd.read_csv(StringIO(data),escapechar='\\')
```

# Read Json to CSV

```python
Data = '{"employee_name": "James", "email": "james@gmail.com", "job_profile": [{"title1":"Team Lead", "title2":"Sr. Developer"}]}'
df1=pd.read_json(Data)
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data', header=None)
```

```python
df.head()
```

```python
#convert json to csv
df.to_csv("wine.csv")
```

```python
#convert jso to different json format
df1.to_json()
```

```python
#orient="records" means we remove index value and print in json format
df1.to_json(orient="records")
```
